```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

def KS(functions, p):
  total_sum = 0
  for func in functions:
    total_sum += np.exp(p * func(x))
  return (1/p) *np.log(total_sum)

# דוגמה לשימוש
f1 = lambda x: -x
f2 = lambda x:  x
#f3 = lambda x:

functions = [f1, f2]
x=np.linspace(-2,2,100)
plt.plot(x,KS(functions,2))
plt.plot(x,np.abs(x))
plt.show()
```
